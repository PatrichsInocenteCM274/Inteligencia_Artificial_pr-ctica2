```python
import pandas as pd
import datetime as dt
```

```python
#Declaramos las rutas para la data
rutaData= 'train.csv'
rutaDataProcesada = 'trainProcesado.csv'

#Declaramos las columnas predictoras y de salida
predictores = ['pickup_datetime','pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','passenger_count']
salida = 'fare_amount'

#La fecha debe tratarse para ser usada en un modelo
FechaHoraInicio="pickup_datetime"

#En predictores, guardaremos la salida para esta primera trata de la data
predictores.append(salida)

#Como prueba, obtenemos el numero de lineas de la data
numeroLineas = sum(1 for row in (open(rutaData)))
print(numeroLineas)
```

```python
#La data no puede cargarse de una sola instancia, por lo que se procesara la misma por chunks
chunksize = 10**5
#Se inicia el loop para 
for i in range(0,numeroLineas,chunksize):
    #Para mantener el header despues de la primera iteracion, se pone un rango de skip sin contar la linea 0
    if i==0:
        df = pd.read_csv(rutaData,nrows = chunksize)
    else:
        df = pd.read_csv(rutaData,nrows = chunksize,skiprows = range(1,i+1))
    #Se procesa las fechas a un valor ordinario
    df[FechaHoraInicio]=pd.to_datetime(df[FechaHoraInicio])
    df[FechaHoraInicio]=df[FechaHoraInicio].map(dt.datetime.toordinal)
    
    #Se elimina las filas con valores 0 en las posciciones
    df=df[df.pickup_longitude !=0]
    df=df[df.pickup_latitude !=0]
    df=df[df.dropoff_longitude !=0]
    df=df[df.dropoff_latitude !=0]
    
    #Se elimina las filas con NaN
    df=df.dropna(axis=0)
    print("bucle completo:",i)
    
    #Igualmente, al guardar la data, no se debe guardar el header despues de la primera iteracion
    if i==0:
        df[predictores].to_csv(rutaDataProcesada,index=False,mode='a',chunksize=chunksize)
    else:
        df[predictores].to_csv(rutaDataProcesada,index=False,header=False,mode='a',chunksize=chunksize)
```
