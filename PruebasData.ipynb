```python
import pandas as pd
import datetime as dt
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
predictores = ['pickup_datetime','pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','passenger_count']
salida = 'fare_amount'
FechaHoraInicio="pickup_datetime"
TamaoChunks = 10 ** 6
for chunk in pd.read_csv('train.csv', chunksize=TamaoChunks):
    chunk[FechaHoraInicio]=pd.to_datetime(chunk[FechaHoraInicio])
    chunk[FechaHoraInicio]=chunk[FechaHoraInicio].map(dt.datetime.toordinal)
    scaler.partial_fit(chunk[predictores])
for chunk in pd.read_csv('train.csv', chunksize=TamaoChunks):
    DataEscalada= scaler.transform(chunk[predictores])
    print(DataEscalada)
```
