```python
#Imports
from sklearn.linear_model import SGDRegressor
import pandas as pd
import datetime as dt
from sklearn.preprocessing import StandardScaler
#Escalamos la data con StandarScaler y usamos SGDRegresor para una prubea de modelo
scaler = StandardScaler()
SGDC = SGDRegressor()
#definimos que se usara para predecir y cual sera el target
predictores = ['pickup_datetime','pickup_longitude','pickup_latitude','dropoff_longitude','dropoff_latitude','passenger_count']
salida = 'fare_amount'
#La fecha debe tratarse para ser usada en un modelo
FechaHoraInicio="pickup_datetime"
#Como la data es muy grande, se trabajara con chunks
TamañoChunks = 10 ** 4
ChunksEntrenar=9
ChunksTotal=ChunksEntrenar+1
for chunk in pd.read_csv('train.csv', chunksize=TamañoChunks, nrows=TamañoChunks*ChunksTotal):
    #Para el trabajo con scaler, se convierte a fecha inicio en un ordinal 
    chunk[FechaHoraInicio]=pd.to_datetime(chunk[FechaHoraInicio])
    chunk[FechaHoraInicio]=chunk[FechaHoraInicio].map(dt.datetime.toordinal)
    scaler.partial_fit(chunk[predictores])
for chunk in pd.read_csv('train.csv', chunksize=TamañoChunks, nrows=TamañoChunks*ChunksEntrenar):
    chunk[FechaHoraInicio]=pd.to_datetime(chunk[FechaHoraInicio])
    chunk[FechaHoraInicio]=chunk[FechaHoraInicio].map(dt.datetime.toordinal)
    #entrenamos el modelo con los chunks a entrenar
    x_train=scaler.transform(chunk[predictores])
    y_train=chunk[salida]
    SGDC.partial_fit(x_train,y_train)
    #print(scaler.transform(chunk[predictores]))
```

```python
from sklearn.metrics import accuracy_score
ChunksUsados=TamañoChunks*ChunksEntrenar
for chunk in pd.read_csv('train.csv', chunksize=TamañoChunks, skiprows=(1,ChunksUsados) ,nrows=TamañoChunks):
    chunk[FechaHoraInicio]=pd.to_datetime(chunk[FechaHoraInicio])
    chunk[FechaHoraInicio]=chunk[FechaHoraInicio].map(dt.datetime.toordinal)
    x_test=scaler.transform(chunk[predictores])
    y_true=chunk[salida]
    y_predict=SGDC.predict(x_test)
    #Calculamos la puntuacion del modelo
    score=SGDC.score(x_test,y_true)
    print(score)
```
