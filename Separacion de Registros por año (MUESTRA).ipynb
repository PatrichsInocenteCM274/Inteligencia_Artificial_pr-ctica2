## SEPARACIÓN DE DATA CON CHUNKS

#### En primer cuando nos proponemos cargar toda la data para un tamaño no soportable de registros, el proceso se facilita mediante el uso de chunks. Seccionando la carga en cantidades mas pequeñas y manejables.

#### En este caso usaremos todo lo que hicimos en este cuaderno de forma condensada y mediante un iterador de chunks

```python
cont=0
for chunk in pd.read_csv('data_ejemplo.csv', chunksize=100):
        chunk['key']=chunk['key'].str.split('-',n=1,expand=True)[0]
        groups_chunk = chunk.groupby(chunk['key'])
        for i in range(len(chunk.key.unique())):
               nombre_csv=chunk.key.unique()[i]+"_con_chunks.csv"
               if cont==0:
                   groups_chunk.get_group(chunk.key.unique()[i]).to_csv(nombre_csv,index=False,mode='a')
               else:
                   groups_chunk.get_group(chunk.key.unique()[i]).to_csv(nombre_csv,index=False,header=False,mode='a')
        cont=cont+1
```

#### Mostremos que se generó:

```python
import os
contenido = os.listdir('/home/adr/my_project_dir/PracticaCalificada2-CC421/separacion')
contenido
```

#### Para podernos asegurar que el proceso con chunks ha generado el mismo resultado, compararemos los archivos comparados con ambos métodos.

```python
import filecmp
print(filecmp.cmp('2009.csv', '2009_con_chunks.csv', shallow=True))  # Si da verdadero significará que son iguales
print(filecmp.cmp('2010.csv', '2010_con_chunks.csv', shallow=True))  # Si da verdadero significará que son iguales
print(filecmp.cmp('2011.csv', '2011_con_chunks.csv', shallow=True)) # Si da verdadero significará que son iguales
print(filecmp.cmp('2012.csv', '2012_con_chunks.csv', shallow=True))  # Si da verdadero significará que son iguales
print(filecmp.cmp('2013.csv', '2013_con_chunks.csv', shallow=True))  # Si da verdadero significará que son iguales
print(filecmp.cmp('2014.csv', '2014_con_chunks.csv', shallow=True))  # Si da verdadero significará que son iguales
print(filecmp.cmp('2015.csv', '2015_con_chunks.csv', shallow=True))  # Si da verdadero significará que son iguales
```
